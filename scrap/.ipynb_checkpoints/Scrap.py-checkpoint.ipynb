```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url="https://www.imdb.com/chart/top/"
```

```python
content = requests.get(url)
```

```python
content
```

```python
html = BeautifulSoup(content.text,"html.parser")
```

```python
soup = html.find_all("tbody")
```

```python
data = soup[0].find_all("tr")
```

```python
len(data)
```

```python
i=1;
with open('innovators.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    print("#,Moives,Year of Release,Rating")
    writer.writerow(["#","Moives","Year of Release","Rating"])
    for x in data:
        temp = x.find_all("a")
        span = x.select('span.secondaryInfo')
        strong = x.select('strong')
        print(str(i)+",\""+temp[1].get_text()+"\","+span[0].get_text()+","+strong[0].get_text())
        writer.writerow([i,temp[1].get_text(),span[0].get_text(),strong[0].get_text()])
        i+=1
    
    
```
